```python
import pandas as pd 
import datetime
import warnings
import numpy as np
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("IIP Data till Dec 21.xlsx")
df.tail(10)
```

```python
dates = df.iloc[4]
date  = [x for x in dates[3:]]
```

```python

data= {"Date":date}
df_new = pd.DataFrame(data)
df_new['Date'] = pd.to_datetime(df_new['Date'])
df_new = df_new.set_index('Date')
df_new = df_new.sort_values(by='Date')
df_new.head()
```

```python
for i in range(5,28):
    new_col =df.loc[i]
    weight = new_col[2]
    res = [pd.to_numeric(x) + weight for x in new_col[3:]]
    col_name = new_col[1]
    df_new[col_name] = res
```

```python
df_new.head()
```

```python
file_name = 'Options_Pricing.xlsx'
# saving the excel
df_new.to_excel(file_name)
```
